```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("track3_obf/DATA/utt_16k.csv")

df2 = pd.read_csv("track3_obf/DATA/utt_24k.csv")

df3 = pd.read_csv("track3_obf/DATA/utt_48k.csv")
```

```python
print(f" size of the dataset 16khz {len(df1)}")
print(f" size of the dataset 24khz {len(df2)}")
print(f" size of the dataset 48khz {len(df3)}")
```

```python
print("=== Dataset Overview ===")
print(f"Total rows: {len(df1)}")
print(f"Columns: {df1.columns.tolist()}")
#print("\nFirst 5 rows:")
#print(df1.head())
#print("\nData types:")
#print(df1.dtypes)
print("\nMissing values per column:")
print(df1.isnull().sum())

# 2. Basic Statistics
print("\n=== Basic Statistics ===")
print("Descriptive statistics for MOS scores:")
print(df1['rating'].describe())
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df1['rating'], kde=True, bins=20)
plt.title('Distribution of MOS Scores 16khz')

plt.subplot(1, 2, 2)
sns.boxplot(x=df1['rating'])
plt.title('Boxplot of MOS Scores 16khz')
plt.tight_layout()
plt.show()

# 5. MOS Scores by ID
plt.figure(figsize=(12, 6))
sns.boxplot(x='lisID', y='rating', data=df1)
plt.title('MOS Scores Distribution by ID 16khz')
plt.xticks(rotation=45)
plt.show()

# 6. MOS Scores Density by ID
plt.figure(figsize=(12, 6))
for id_val in df1['lisID'].unique():
    sns.kdeplot(df1[df1['lisID'] == id_val]['rating'], label=f'ID {id_val}')
plt.title('Density of MOS Scores by ID 16khz')
plt.xlabel('MOS Score')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df2['rating'], kde=True, bins=20)
plt.title('Distribution of MOS Scores 24khz')

plt.subplot(1, 2, 2)
sns.boxplot(x=df2['rating'])
plt.title('Boxplot of MOS Scores 24khz')
plt.tight_layout()
plt.show()

# 5. MOS Scores by ID
plt.figure(figsize=(12, 6))
sns.boxplot(x='lisID', y='rating', data=df2)
plt.title('MOS Scores Distribution by ID 24khz')
plt.xticks(rotation=45)
plt.show()

# 6. MOS Scores Density by ID
plt.figure(figsize=(12, 6))
for id_val in df2['lisID'].unique():
    sns.kdeplot(df2[df2['lisID'] == id_val]['rating'], label=f'ID {id_val}')
plt.title('Density of MOS Scores by ID 24khz')
plt.xlabel('MOS Score')
plt.legend()
plt.show()
```
