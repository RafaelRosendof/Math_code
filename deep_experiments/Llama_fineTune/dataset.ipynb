```python
df1 = pd.read_csv('dadosPLN/portuguese-poems.csv')
df2 = pd.read_csv('dadosPLN/controle_armas.csv')
df3 = pd.read_csv('dadosPLN/vacinacao.csv')
df4 = pd.read_csv('dadosPLN/violencia_policial.csv')
df5 = pd.read_csv('dadosPLN/legalizacao_aborto.csv')
df6 = pd.read_csv('dadosPLN/maioridade_penal.csv')
df7 = pd.read_csv('dadosPLN/mudancas_climaticas.csv')
```

```python
df1['Author'] = 'Poema de ' + df1['Author']
df1['Views'] = '\n\n'
df1.drop(columns=['Views'], inplace=True)
df1['Title'] = 'O título da obra é: ' + df1['Title']
df1['Content'] = 'O poema é: ' + df1['Content'] + '\n\n'

df1.to_csv('portuguese-poems2.csv', index=False)
```

```python
df2['question'] = 'Pergunta sobre o tema: ' + df2['question']
df2['chatgpt'] = 'Resposta do ChatGPT: ' + df2['chatgpt']
df2['chatgpt_answer_type'] = 'Tipo de resposta do ChatGPT: ' + df2['chatgpt_answer_type']
df2['bard'] = 'Resposta do Bard: ' + df2['bard']
df2['bard_answer_type'] = 'Tipo de resposta do Bard: ' + df2['bard_answer_type']
df2['bing'] = 'Resposta do Bing: ' + df2['bing']
df2['bing_answer_type'] = 'Tipo de resposta do Bing: ' + df2['bing_answer_type']
df2['cabrallama'] = 'Resposta do Cabrallama: ' + df2['cabrallama']
df2['cabrallama_answer_type'] = 'Tipo de resposta do Cabrallama: ' + df2['cabrallama_answer_type']
```

```python
def trata_csv(df , nome_arquivo):
    df['question'] = 'Pergunta sobre o tema: ' + df['question'] 
    df['chatgpt'] = '\nResposta do ChatGPT: ' + df['chatgpt']
    df['chatgpt_answer_type'] = '\nTipo de resposta do ChatGPT: ' + df['chatgpt_answer_type']
    df['bard'] = '\nResposta do Bard: ' + df['bard']
    df['bard_answer_type'] = '\nTipo de resposta do Bard: ' + df['bard_answer_type']
    df['bing'] = '\nResposta do Bing: ' + df['bing']
    df['bing_answer_type'] = '\nTipo de resposta do Bing: ' + df['bing_answer_type']
    df['cabrallama'] = '\nResposta do Cabrallama: ' + df['cabrallama']
    df['cabrallama_answer_type'] = '\nTipo de resposta do Cabrallama: ' + df['cabrallama_answer_type']
    df.to_csv(nome_arquivo, index=False)

trata_csv(df2, 'dadosPLN/controle_armas2.csv')
trata_csv(df3, 'dadosPLN/vacinacao2.csv')
trata_csv(df4, 'dadosPLN/violencia_policial2.csv')
```

```python
trata_csv(df5, 'dadosPLN/legalizacao_aborto2.csv')
trata_csv(df6, 'dadosPLN/maioridade_penal2.csv')
trata_csv(df7, 'dadosPLN/mudancas_climaticas2.csv')
```

```python
path = 'dadosPLN/corpus/'

print(f"Verificando arquivos no diretório: '{path}'\n")

# Lista todos os arquivos e diretórios na pasta especificada
for filename in os.listdir(path):
    
    # 1. Condição: Verifica se o nome do arquivo contém um espaço
    if ' ' in filename:
        
        # Constrói o caminho completo para o arquivo original
        # Ex: 'dadosPLN/corpus/Os Romances da Semana.txt'
        old_filepath = os.path.join(path, filename)
        
        # 2. Medida de segurança: Garante que é realmente um arquivo (e não uma pasta)
        if os.path.isfile(old_filepath):
            
            # 3. Cria o novo nome do arquivo substituindo espaços por underscores
            # Ex: 'Os_Romances_da_Semana.txt'
            new_filename = filename.replace(' ', '_')
            
            # Constrói o caminho completo para o novo nome do arquivo
            new_filepath = os.path.join(path, new_filename)
            
            # Imprime a ação que será realizada para o usuário ver
            print(f"Renomeando: '{filename}'  ->  '{new_filename}'")
            
            # 4. Renomeia o arquivo no sistema
            os.rename(old_filepath, new_filepath)

print("\nProcesso de renomeação concluído!")
```
