```python
import jax 
import jax.numpy as jnp
```

```python
def f(x):
    return x**2 
```

```python
h = 1e-12


def derivate(f ,x):
    derivada = (f (x +h) - f(x)) / h
    return derivada
```

```python
print(derivate(f, 1000))
```

```python
def f_sin(x):
    return jnp.sin(x)
def f_cos(x):
    return jnp.cos(x)
```

```python
def derivate2(f_sin , x):
    #derivada = jax.grad(f_sin)(x)
    derivada = (f_sin(x + h) - f_sin(x)) / h
    return derivada

def derivate3(f_cos , x):
    derivada = (f_cos(x + h) - f_cos(x)) / h
    return derivada
```

```python
print(derivate2(f_sin, 90))
print(derivate3(f_cos, 90))
```

```python
h = 0.0001

def calcular_derivada_aproximada(funcao, x):
    derivada = (funcao(x + h) - funcao(x)) / h
    return derivada

def f_sin(x):
    return jnp.sin(x)

def f_cos(x):
    return jnp.cos(x)


angulo_rad = jnp.pi 

derivada_seno = calcular_derivada_aproximada(f_sin, angulo_rad)
print(f"A derivada aproximada de sen(90°) é: {derivada_seno:.5f}") 


derivada_cosseno = calcular_derivada_aproximada(f_cos, angulo_rad)
print(f"A derivada aproximada de cos(90°) é: {derivada_cosseno:.5f}") #

derivada_sin = calcular_derivada_aproximada(f_cos , jnp.pi/3)
print(f"A derivada aproximada de cos(60°) é: {derivada_sin:.5f}")

derivada_sin = calcular_derivada_aproximada(f_sin , jnp.pi/3)
print(f"A derivada aproximada de sin(60°) é: {derivada_sin:.5f}")
```
