Notebook for calculating expectation

```python
import numpy as np 
import scipy as sp
import matplotlib.pyplot as plt
```

```python
### first we gonna build a CFD and a PDF 

vec = np.linspace(1, 100, 10**6)

vec_2 = np.random.beta(1, 10, vec.shape)

vec_2
```

```python
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(vec_2)
```

```python
### calculating the area

exp = sp.integrate.trapezoid(vec_2)

exp
```

```python
### calculating the expectation 

f = lambda x: x * sp.stats.beta.pdf(x, 1, 100)

exp = sp.integrate.quad(f, 0, 100)

exp
```
