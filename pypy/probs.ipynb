```python
nvidia = yf.Ticker("ORCL") 
print(nvidia.history(period="1d")['Close'].iloc[-1])

#print(nvidia.get_dividends())
#print(nvidia.history())
#print(nvidia.options())
#print(nvidia.cash_flow())
#print(nvidia.earnings_dates())
#print(nvidia.earnings())

stock = yf.Ticker("AAPL")
df = stock.history(period="1m", interval="1d", auto_adjust=True)
df['Retorno'] = df['Close'].pct_change()
df['Log_Retorno'] = np.log(df['Close'] / df['Close'].shift(1))

vol = df['Log_Retorno'].std() * np.sqrt(30)

try:
    beta = stock.info.get('beta')
    pe_ratio = stock.info.get('forward')
    print(f"Beta (market_risky) {beta}")
    print(f"P/L Projecting: {pe_ratio}")
except Exception as e:
    print(e)
    
if stock.options:
    vec = stock.options[0]
    chain = stock.option_chain(vec)
    iv_av = chain.calls['impliedVolatility'].mean()
    
    print(chain.calls)
    
    print(f"Volatilidade Implicita (Opção {vec}): {iv_av:.2%}")
```
